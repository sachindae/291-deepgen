```python
import pickle
import matplotlib.pyplot as plt
```

```python
# metrics
file_name = "./Lyrics2Melody/Lyrics-Conditioned-Neural-Melody-Generation/data/metric.pkl"

with open(file_name, 'rb') as f:
    metrics = pickle.load(f)
print(len(metrics['mse']))
```

```python
#stats
file_name = "./Lyrics2Melody/Lyrics-Conditioned-Neural-Melody-Generation/data/stats.pkl"

with open(file_name, 'rb') as f:
    stats = pickle.load(f)
print(len(stats))
```

```python
# metric graph
for key in metrics:
    plt.plot(metrics[key])
    plt.xlabel("epoch")
    plt.ylabel(key)
    plt.savefig('images/'+str(key)+'.png')
    plt.show()
```

```python
#stats deform
stats_df = {}
for key in stats[0]:
    stats_df[key] = []
for stat in stats:
    for key in stat:
        stats_df[key].append(stat[key])
```

```python
# stats graph
for key in stats_df:
    plt.plot(stats_df[key])
    plt.xlabel("epoch")
    plt.ylabel(key)
    plt.savefig('images/'+str(key)+'.png')
    plt.show()
```
