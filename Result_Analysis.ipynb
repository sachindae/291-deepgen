```python
import pickle
import matplotlib.pyplot as plt
```

```python
# metrics
file_name = "./Lyrics2Melody/Lyrics-Conditioned-Neural-Melody-Generation/data/metric.pkl"

with open(file_name, 'rb') as f:
    metrics = pickle.load(f)
print(len(metrics['mse']))
```

```python
#stats
file_name = "./Lyrics2Melody/Lyrics-Conditioned-Neural-Melody-Generation/data/stats.pkl"

with open(file_name, 'rb') as f:
    stats = pickle.load(f)
print(len(stats))
```

```python
# metric graph
for key in metrics:
    plt.plot(metrics[key])
    plt.xlabel("epoch")
    plt.ylabel(key)
    plt.savefig('images/'+str(key)+'.png')
    plt.show()
```

```python
#stats deform
stats_df = {}
for key in stats[0]:
    stats_df[key] = []
for stat in stats:
    for key in stat:
        stats_df[key].append(stat[key])
```

```python
# print(f"MIDI numbers span: {stat['stats_span_tot']}")
# print(f"3-MIDI-number repetitions: {stat['stats_repetitions_3_tot']}")
# print(f"2-MIDI-number repetitions: {stat['stats_repetitions_2_tot']}")
# print(f"# Unique MIDI numbers: {stat['stats_unique_tones_tot']}")
# print(f"# Notes without rest: {stat['num_of_null_rest_tot']}")
# print(f"Avg rest value within a song: {stat['stats_avg_rest_tot']}")
# print(f"Song length(average): {stat['avg_songlength']}")
# print(f"Number of Good Songs: {stat['num_good_songs']}")
```

```python
# stats graph
for key in stats_df:
    if key == "stats_span_tot":
      label = "MIDI numbers span"
    elif key == "stats_repetitions_3_tot":
      label = "3-MIDI-number repetitions"
    elif key == "stats_repetitions_2_tot":
      label = "2-MIDI-number repetitions"
    elif key == "stats_unique_tones_tot":
      label = "# Unique MIDI numbers"
    elif key == "num_of_null_rest_tot":
      label = "# Notes without rest"
    elif key == "stats_avg_rest_tot":
      label = "Avg rest value within a song"
    elif key == "avg_songlength":
      label = "Song length(average)"
    elif key == "num_good_songs":
      label = "# Songs with typical range"
    else:
      label = key
    plt.plot(stats_df[key])
    plt.xlabel("epoch")
    plt.ylabel(label)
    plt.savefig('images/'+str(label)+'.png')
    plt.show()
```

```python
def lyricsclass(lyr):
    lyr_s = lyr.split(" ")
    rets = []
    ret = []
    i=0
    for l in lyr_s:
        ret.append([l,l])
        if i>20:
            rets.append(ret)
            ret = []
            i=0
        i += 1
    return rets
```
